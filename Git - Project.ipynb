## Scope and key terms

##### GIT : Git is a open-source version control system for management of code. Git is very popularly used for small to very large projects
    
  
    

Repository : Repo has all the directories contents of the project.<br>
Branch : Branch is the alternate version of the actual repo, which helps in making experimental code changes<br>
Push : Push command helps in pushing the code changes to the remote repository<br>
Pull : Pull command helps in pushing the code changes from another branch to main branch after review by the repo maintainers<br>
Clone : Cloning is the process of making a copy of existing repository<br>
Fork : Forking does copy the exisiting repository and creates a complete independent repo from the actual one.
